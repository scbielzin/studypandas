```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
  'Estudo': [128, 73, 77],
  'Minutos': [45, 32, 67]
}

#rotulos são index padrão 0, 1, 2

myvar = pd.DataFrame(data, index = ["dia1", "dia2", "dia3"])

myvar.plot()

plt.show()

#Pandas usam o loc atributo para retornar uma ou mais linhas especificadas
```
